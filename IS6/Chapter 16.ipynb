```python
# Plot between -4 and 4 with .001 steps.
x_axis = np.arange(-4, 4, 0.001)
  
plt.plot(x_axis, norm.pdf(x_axis))
plt.vlines(-0.08, 0, 0.45)
plt.show()
```

#### Section 16.2: Alpha Levels and Critical Values

```python
#figure 16.1, page 543
x_axis = np.arange(-4,4,0.001)

plt.plot(x_axis, norm.pdf(x_axis))
plt.vlines([-1.96, 1.96], 0, 0.4)
plt.show()
```
