### Chapter 4: Understanding and Comparing Distributions

```python
#Read in libraries. These are all the libraries we need for this chapter
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Figure 4.1, page 99
forest = pd.read_csv("http://nhorton.people.amherst.edu/is5/data/Hopkins_Forest.csv", sep = ",")

sns.histplot(data = forest, x = "Avg Wind (mph)")
plt.xlabel("Average Wind Speed (mph)")
plt.ylabel("# of Days")
plt.show()

print(forest["Avg Wind (mph)"].describe())
```

#### Section 4.1: Displays for Comparing Groups

```python
#Figure 4.2, page 99
#Wrangle data
season = []
for i in range(len(forest)):
    if forest["Month"][i] <= 9 and forest["Month"][i] >= 4:
        season.append("Spring/Summer")
    else:
        season.append("Fall/Winter")
forest["Season"] = season

#Plot
g = sns.FacetGrid(data = forest, col = "Season")
g.map(sns.histplot, "Avg Wind (mph)")

plt.show()

print(forest.groupby("Season")["Avg Wind (mph)"].describe())
```

#### Boxplots

```python
# Replicating a single boxplot from page 101
# Step 4 on page 101
forest = pd.read_csv("http://nhorton.people.amherst.edu/is5/data/Hopkins_Forest.csv", sep = ",")

sns.boxplot(data = forest, x = "Avg Wind (mph)")
plt.ylabel("Daily Average Wind Speed (mph)")
plt.show()
```

```python
#Side by side boxplots, figure 4.3, page 102

sns.boxplot(data = forest, x = "Month", y = "Avg Wind (mph)", orient = "v", color = "green")
plt.show()
```

#### Step-by-step Example: Comparing groups

```python
#Mechanics, page 103
cups = pd.read_csv("datasets/cups.txt", sep = "\t")

print(cups.groupby("Container").describe())

sns.boxplot(data = cups, x = "Container", y = "Difference")
plt.ylabel("Temp Change (F)")
plt.show()
```
