```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

#### Section 8.1: Examining Residuals
#### Getting the "Bends": When the residuals aren't straight

```python
penguins = pd.read_csv("datasets/penguins.txt", sep = "\t")
penguins

#Figure 8.1, page 248
sns.scatterplot(data = penguins, x = "Duration(min)", y = "Dive Heart Rate", color = "brown")
plt.xlabel("Duration (min)")
plt.ylabel("Dive Heart Rate (bpm)")
plt.show()
```

```python
#Figure 8.2
#Fit linear regression model
x = penguins["Duration(min)"]
y = penguins["Dive Heart Rate"]
x = sm.add_constant(x)
model = sm.OLS(y,x)
result = model.fit()

#Plot residuals against duration
sns.scatterplot(x = penguins["Duration(min)"], y = result.resid, color = "green")
plt.xlabel("Duration (min)")
plt.ylabel("Residuals")
plt.show()
```

#### Sifting Residuals for Groups

```python
#Figure 8.3, page 249
cereals = pd.read_csv("datasets/cereals.txt", sep = "\t")
cereals

#Fit linear model
x = cereals["sugars"]
y = cereals["calories"]
x = sm.add_constant(x)
model = sm.OLS(y,x)
result = model.fit()

#Plot the histogram of residuals
sns.histplot(x = result.resid, color = "green", binwidth = 8)
plt.xlabel("Residuals (calories)")
plt.ylabel("# of Cereals")
plt.show()
```

#### Section 8.3: Outliers, Leverage, and Influence

```python
#Figure 8.10, page 255
election = pd.read_csv("datasets/election-2000.txt", sep = "\t")

#Make linear model
x = sm.add_constant(election["Nader"])
y = election["Buchanan"]
model = sm.OLS(y,x)
result = model.fit()

#Summary
print(result.summary())
```

```python
#Make linear model
data = election[election["Buchanan"] <= 3000]
x = sm.add_constant(data["Nader"])
y = data["Buchanan"]
model = sm.OLS(y,x)
result = model.fit()

#Summary
print(result.summary())
```

```python
#Figure 8.10, page 255
sns.lmplot(data = election, x = "Nader", y = "Buchanan", ci = None)
sns.lmplot(data = data, x = "Nader", y = "Buchanan", ci = None)
plt.show()
```

#### Example 8.2: Using Several of These Methods Together

```python
#page 259
bikes = pd.read_csv("datasets/dirt-bikes-2014.txt", sep = "\t")
bikes

sns.scatterplot(data = bikes, x = "Displacement", y = "MSRP")
plt.show()
```

```python
#Fit linear model
x = sm.add_constant(bikes["Displacement"])
y = bikes["MSRP"]
model = sm.OLS(y,x)
result = model.fit()

#Make predictions
yhat = result.fittedvalues
residuals = result.resid

#Plot
sns.scatterplot(x = yhat, y = residuals)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.show()
```

```python
#Mutate Displacement column
bikes = bikes.assign(new_dist = bikes["Displacement"] ** (1/3))

#Plot linear regression
sns.lmplot(data = bikes, x = "new_dist", y = "MSRP", ci = None, hue = "Cooling", hue_order = ["Liquid-Cooled", "Air-Cooled"])
plt.xlabel("(Displacement) ^(1/3)")
plt.show()
```

#### Section 8.6: Straightening Scatterplots - The Three Goals

```python
fuel = pd.read_csv("datasets/fuel-efficiency.txt", sep = "\t")
fuel

sns.scatterplot(data = fuel[fuel["City Mpg"] <= 40], x = "Weight", y = "City Mpg")
plt.xlabel("Weight (lb)")
plt.ylabel("Fuel Efficiency (mpg)")
plt.show()
```

```python
#Fit linear model
data = fuel[fuel["City Mpg"] <= 40]
x = sm.add_constant(data["Weight"])
y = data["City Mpg"]
model = sm.OLS(y,x)
result = model.fit()

#Make predictions
yhat = result.fittedvalues
residuals = result.resid

#Plot
sns.scatterplot(x = yhat, y = residuals)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.show()
```

```python
#Mutate dataframe; add new column
fuel = fuel.assign(fuel_consumption = (1 / fuel["City Mpg"]) * 100)

sns.scatterplot(data = fuel, x = "Weight", y = "fuel_consumption")
plt.show()
```

```python
#Fit linear model
x = sm.add_constant(fuel["Weight"])
y = fuel["fuel_consumption"]
model = sm.OLS(y,x)
result = model.fit()

#Make predictions
yhat = result.fittedvalues
residuals = result.resid

#Plot
sns.scatterplot(x = yhat, y = residuals)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.show()
```

#### Goals of Re-Expression for Regressio
We use the hand dexterity data to illustrate re-expressions

```python
hand = pd.read_csv("datasets/hand-dexterity.txt", sep = "\t")
hand

sns.scatterplot(data = hand, x = "Age(yr)", y = "Time(sec)", hue = "Dominant")
plt.show()
```

```python
hand = hand.assign(speed = 1 / hand["Time(sec)"])

sns.lmplot(data = hand, x = "Age(yr)", y = "Speed", hue = "Dominant", ci = None)
plt.show()
```

```python
#Fit linear model
hand = hand.dropna()
x = sm.add_constant(hand["Age(yr)"])
y = hand["Speed"]
model = sm.OLS(y,x).fit()

#Plot
sns.histplot(x = model.resid, bins = 7)
plt.xlabel("Residuals")
plt.show()
```

#### Section 8.7: Finding a Good Re-expression
See the table 8.1 and Figure 8.24 on page 264 - 265
#### Step-By-Step Example: Re-Expressing to Straighten a Scatterplot
We can explore different re-expressions

```python
penguins = pd.read_csv("datasets/penguins.txt", sep = "\t")
penguins
sns.scatterplot(data = penguins, x = "Duration(min)", y = "Dive Heart Rate")
plt.xlabel("Duration")
plt.ylabel("DHR")
plt.show()
```

```python
#Fit linear model
x = sm.add_constant(penguins["Duration(min)"])
y = penguins["Dive Heart Rate"]
model = sm.OLS(y,x).fit()

#Plot
sns.scatterplot(x = model.fittedvalues, y = model.resid)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.show()
```

```python
sns.histplot(x = model.resid, binwidth = 10)
plt.xlabel("Residuals")
plt.ylabel("# of Residuals")
plt.show()
```

```python
#Mutating with the square root
sns.scatterplot(data = penguins, x = "Duration(min)", y = penguins["Dive Heart Rate"] ** (1/2))
plt.xlabel("Duration")
plt.ylabel("sqrt(DHR)")
plt.show()
```

```python
sns.scatterplot(data = penguins, x = "Duration(min)", y = np.log(penguins["Dive Heart Rate"]))
plt.xlabel("Duration")
plt.ylabel("Log(DHR)")
plt.show()
```

```python
#Fit new linear model
x = sm.add_constant(penguins["Duration(min)"])
y = np.log(penguins["Dive Heart Rate"])
model = sm.OLS(y,x).fit()

#Plot
sns.scatterplot(x = model.fittedvalues, y = model.resid)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.show()
```

```python
sns.histplot(x = model.resid, binwidth = 0.1)
plt.xlabel("Residuals")
plt.ylabel("# of Residuals")
plt.show()
```

```python
#Mutating with a (negative) reciprocal
sns.scatterplot(data = penguins, x = "Duration(min)", y = (-1 / penguins["Dive Heart Rate"]))
plt.xlabel("Duration")
plt.ylabel("-1/DHR")
plt.show()
```

```python
#Fit linear model
x = sm.add_constant(penguins["Duration(min)"])
y = -1 / penguins["Dive Heart Rate"]
model = sm.OLS(y,x).fit()

#Plot
sns.scatterplot(x = model.fittedvalues, y = model.resid)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.show()
```

```python
sns.histplot(x = model.resid, binwidth = 0.0025)
plt.xlabel("Residuals")
plt.ylabel("# of Residuals")
plt.show()
```
