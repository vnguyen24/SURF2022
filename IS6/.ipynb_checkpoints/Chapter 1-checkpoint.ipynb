##### df.iloc explanation:

iloc lets you specify the range of rows/columns you want to output. The full syntax is **df.iloc[starting_row:ending_row,starting_column:ending_column]**. Note that ending rows and columns are exclusive (e.g: ending row at index 12 means everything up to index 11 will be included, but not index 12)
